```python
import sys
import requests
import json
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 3)
```

## 1. Position data for time interval

```python
bodies = ['sun', 'earth', 'jupiter', 'saturn', 'neptune', 'uranus', 'venus', 'mars', 'mercury']
d = []
output_file = "horizons_data_1750_250_years_to_plot_1_day_timestep.csv"
folder = "batch_files/to_plot/"
```

```python
for body in bodies:
    input_file = folder + "file_%s.txt"%(body)
    f = open(input_file)
    url = 'https://ssd.jpl.nasa.gov/api/horizons_file.api'
    r = requests.post(url, data={'format':'json'}, files={'input': f})
    f.close()
#     print(r.text)
    
    data = json.loads(r.text)
    data = data['result'].split('\n')
    
    print(len(data))
    for i in range(len(data)):
        if data[i] == '$$SOE':
            i_start = i+1
        if data[i] == '$$EOE':
            i_end = i

    data = data[i_start: i_end]
#     print(data)
    d.append(data)
```

```python
print(len(d))
tmp_list = ['X', 'Y', 'Z', '=', '', 'VX', 'VY', 'VZ', 'VX=', 'VY=', 'VZ=']
df = pd.DataFrame()

for n in range(len(bodies)):
    positions = []
    velocities = []
    data = d[n]
    for i in range(len(data)):
        if i % 4 == 1:
            s = ""
#             print(data[i])
            tmp = data[i].split(" ")
            for elem in tmp:
                if elem not in tmp_list:
                    elem = elem.replace("=", "")
                    elem = elem.replace("", "")
                    if s == "":
                        s = s + str(float(elem))
                    else:
                        s = s + "," + str(float(elem))
            positions.append(s)
        if i % 4 == 2:
            s = ""
#             print(data[i])
            tmp = data[i].split(" ")
            for elem in tmp:
                if elem not in tmp_list:
                    elem = elem.replace("=", "")
                    elem = elem.replace("", "")
                    elem = elem.replace("VY", "")
                    elem = elem.replace("VX", "")
                    elem = elem.replace("VZ", "")
                    if s == "":
                        s = s + str(float(elem))
                    else:
                        s = s + "," + str(float(elem))
            velocities.append(s)
            
    # print(positions)
    # print(velocities)
    df['position_%s'%(bodies[n])] = positions
    df['velocity_%s'%(bodies[n])] = velocities
```

```python
display(df)
```

```python
df.to_csv(output_file, index = False)
```
